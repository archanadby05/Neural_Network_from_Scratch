## Tanh Neuron - Activation Centered at 0

### **01. Define the Tanh Function and Its Derivative**

Tanh squashes input values into the range (-1, 1), making it zero-centered. This helps gradients flow better during training in some cases.


```python
import numpy as np
import matplotlib.pyplot as plt

# Tanh activation and its derivative
def tanh(x):
    return np.tanh(x)

def tanh_derivative(x):
    return 1 - np.tanh(x) ** 2
```

*Explanation:*

Unlike sigmoid, which outputs in (0, 1), tanh outputs between -1 and 1, making it symmetric around zero. This is often helpful in reducing bias in updates during training.

### **02. Define TanhNeuron Class**

We build a neuron that performs a forward pass using the tanh activation function.

```python
class TanhNeuron:
    def __init__(self, input_dim):
        self.weights = np.zeros(input_dim)
        self.bias = 0

    def forward(self, x):
        z = np.dot(x, self.weights) + self.bias
        return tanh(z)
```

*Explanation:*

The class uses the tanh activation after computing a weighted sum of the inputs.

### **03. Manual Forward Pass**

We create a TanhNeuron and pass in inputs to compute the output.

```python
inputs = np.array([1.0, -1.5])
neuron = TanhNeuron(input_dim=2)
output = neuron.forward(inputs)

print("Tanh neuron output:", output)
```

*Explanation:*

This tests the neuron on a simple input and prints the activated output using tanh.

### **04. Plot Activation Curve (tanh vs sigmoid)**

We visualize the tanh and sigmoid activation functions to understand their shape and differences.

```python
# Sigmoid for comparison
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x_vals = np.linspace(-5, 5, 200)
tanh_vals = tanh(x_vals)
sigmoid_vals = sigmoid(x_vals)

plt.figure(figsize=(8, 5))
plt.plot(x_vals, tanh_vals, label="tanh", color="red")
plt.plot(x_vals, sigmoid_vals, label="sigmoid", color="blue", linestyle="--")
plt.axhline(0, color='gray', linewidth=0.5)
plt.title("Activation Function Comparison: tanh vs sigmoid")
plt.xlabel("Input (z)")
plt.ylabel("Activation Output")
plt.grid(True)
plt.legend()
plt.show()
```

*Explanation:*

This plot shows that while sigmoid outputs between 0 and 1, tanh outputs between -1 and 1. Tanh is zero-centered, which can be beneficial for certain tasks.
