## ReLU Activation and Comparison with Sigmoid

### 01. ReLU Function and Derivative

We define the Rectified Linear Unit (ReLU) activation function and its derivative.

```python
import numpy as np
import matplotlib.pyplot as plt

def relu(x):
    return np.maximum(0, x)

def relu_derivative(x):
    return np.where(x > 0, 1, 0)
```

*Explanation:*

ReLU outputs 0 for negative inputs and passes positive values unchanged. It's simple, fast, and avoids vanishing gradients in positive ranges. Its derivative is 1 for positive inputs, 0 otherwise.

### 02. Plot ReLU Curve

We visualize how ReLU behaves across input values.

```python
x = np.linspace(-10, 10, 400)
y = relu(x)

plt.plot(x, y, label='ReLU', color='green')
plt.title("ReLU Activation Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.legend()
plt.show()
```

*Explanation:*

The ReLU curve is flat for negative inputs and linear for positive ones. This allows neurons to activate only when necessary, creating sparse representations.

### 03. Sparsity in Activations

*Explanation:*

ReLU introduces sparsity, as many outputs become zero (especially early in training). This helps reduce interdependencies between neurons, leading to better generalization and efficient computation.

### 04. Compare with Sigmoid

We compare ReLU and sigmoid side by side.

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

y_sigmoid = sigmoid(x)

plt.plot(x, relu(x), label='ReLU', color='green')
plt.plot(x, y_sigmoid, label='Sigmoid', color='blue')
plt.title("ReLU vs Sigmoid Activation")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.legend()
plt.axhline(0, color='gray', linestyle='--', linewidth=0.5)
plt.axvline(0, color='gray', linestyle='--', linewidth=0.5)
plt.show()
```

*Explanation:*

While sigmoid saturates and slows training due to vanishing gradients, ReLU stays linear in the positive region, allowing faster convergence and deeper networks.

### 05. Simple ReLU Neural Layer Demo

We simulate a one-layer ReLU network to illustrate sparsity in output.

```python
np.random.seed(42)

# Simulated input and weights
inputs = np.random.randn(5, 3)  # 5 samples, 3 features
weights = np.random.randn(3, 4)  # 3 inputs → 4 neurons
biases = np.random.randn(1, 4)

# Linear combination + ReLU
z = np.dot(inputs, weights) + biases
a = relu(z)

print("ReLU Activations:\n", a)
```

*Explanation:*

This example shows many values in the activation matrix are zero due to ReLU, confirming its sparse output behavior.
