## Tanh Activation and Comparison with Sigmoid

### 01. Tanh Function and Derivative

We define the hyperbolic tangent (tanh) activation function and its derivative.

```python
import numpy as np
import matplotlib.pyplot as plt

def tanh(x):
    return np.tanh(x)

def tanh_derivative(x):
    return 1 - np.tanh(x) ** 2
```

*Explanation:*

Tanh outputs values between -1 and 1. It's zero-centered, which can help the network converge faster than sigmoid in some cases. Its derivative is important for gradient-based optimization.

### 02. Plot Tanh Curve

We visualize how the tanh function responds to input values.

```python
x = np.linspace(-10, 10, 400)
y = tanh(x)

plt.plot(x, y, label='tanh', color='red')
plt.title("Tanh Activation Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.legend()
plt.show()
```

*Explanation:*

The tanh curve is S-shaped like sigmoid, but spans from -1 to 1. It's symmetric around 0, which can lead to better gradient flow during training.


### 03. Compare with Sigmoid Function

We plot both tanh and sigmoid functions to highlight their differences.

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

y_sigmoid = sigmoid(x)

plt.plot(x, tanh(x), label='tanh', color='red')
plt.plot(x, y_sigmoid, label='sigmoid', color='blue')
plt.title("Tanh vs Sigmoid Activation")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.legend()
plt.axhline(0, color='gray', linestyle='--', linewidth=0.5)
plt.axvline(0, color='gray', linestyle='--', linewidth=0.5)
plt.show()
```

*Explanation:*

Unlike sigmoid, which outputs between 0 and 1, tanh is centered around 0. This helps reduce bias shifts in activations and can improve training dynamics.

### 04. Why Tanh Can Be Better

*Explanation:*

Since tanh outputs both negative and positive values, it helps maintain a more balanced gradient flow across layers. This often leads to faster convergence than sigmoid in deep networks, especially when inputs are normalized.
