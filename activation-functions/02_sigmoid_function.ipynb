## Sigmoid Activation and Comparison with Tanh

### 01. Sigmoid Function and Derivative

We define the sigmoid activation function and its derivative, commonly used in neural networks.

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    sx = sigmoid(x)
    return sx * (1 - sx)
```

*Explanation:*

The sigmoid squashes inputs into the range (0, 1), making it useful for binary outputs. Its derivative is used in backpropagation to update weights.



### 02. Plot Sigmoid Curve

We visualize how the sigmoid function behaves across a range of values.

```python
x = np.linspace(-10, 10, 400)
y = sigmoid(x)

plt.plot(x, y, label='sigmoid', color='blue')
plt.title("Sigmoid Activation Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.legend()
plt.show()
```

*Explanation:*

The curve smoothly transitions from 0 to 1. Around 0, the function is steepest, meaning it's most responsive to input changes.

### 03. Compare with Tanh Function

We plot both sigmoid and tanh functions for comparison.

```python
def tanh(x):
    return np.tanh(x)

y_tanh = tanh(x)

plt.plot(x, sigmoid(x), label='sigmoid', color='blue')
plt.plot(x, y_tanh, label='tanh', color='red')
plt.title("Sigmoid vs Tanh Activation")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.legend()
plt.axhline(0, color='gray', linestyle='--', linewidth=0.5)
plt.axvline(0, color='gray', linestyle='--', linewidth=0.5)
plt.show()
```

*Explanation:*

Tanh is zero-centered and ranges from -1 to 1. This can make optimization easier in some cases, especially in deeper networks.

### 04. Vanishing Gradient Discussion

In both sigmoid and tanh, the gradient becomes very small at the tails (large positive or negative inputs). This leads to the vanishing gradient problem, where weights update very slowly, hindering deep network training.
