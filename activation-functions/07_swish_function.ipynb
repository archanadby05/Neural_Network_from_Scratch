## Swish Activation Function and Comparison with Other Activations

### 01. Swish Function Implementation

We define the Swish activation function: swish(x) = x * sigmoid(x).

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def swish(x):
    return x * sigmoid(x)
```

*Explanation:*

Swish is a smooth, non-monotonic activation function introduced by Google. It retains benefits of ReLU while being differentiable everywhere. Unlike ReLU, Swish allows small negative outputs, improving expressiveness.

### 02. Visualize Swish Curve


We plot the Swish function to observe its shape and behavior.

```python
x = np.linspace(-10, 10, 400)
y_swish = swish(x)

plt.plot(x, y_swish, label="Swish", color="purple")
plt.title("Swish Activation Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.legend()
plt.show()
```

Explanation:
Swish behaves linearly for large positive values, suppresses large negatives softly, and allows gradients to flow through small negative values—unlike ReLU which kills them completely.

### 03. Compare Swish with ReLU, Sigmoid, and Tanh

We visualize how Swish differs from standard activations.

```python
def relu(x):
    return np.maximum(0, x)

def tanh(x):
    return np.tanh(x)

y_relu = relu(x)
y_sigmoid = sigmoid(x)
y_tanh = tanh(x)

plt.plot(x, swish(x), label="Swish", color="purple", linewidth=2)
plt.plot(x, y_relu, label="ReLU", linestyle="--", color="green")
plt.plot(x, y_sigmoid, label="Sigmoid", linestyle="--", color="blue")
plt.plot(x, y_tanh, label="Tanh", linestyle="--", color="orange")
plt.title("Swish vs Other Activations")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.legend()
plt.show()
```

*Explanation:*

ReLU kills negatives completely.

Sigmoid and tanh squash values but suffer from vanishing gradients.

Swish combines benefits of smoothness, nonlinearity, and gradient flow across all input ranges.


### 04. When to Use Swish

*Explanation:*

Swish often outperforms ReLU in deep networks and modern architectures (e.g., EfficientNet). It’s especially helpful when:

- Training very deep models

- Needing smooth gradient flow

- Allowing mild negative activations improves performance
